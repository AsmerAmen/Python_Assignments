Write a program that simulates a Civil Registry using csv files which enables the user to do the following:
    - A user can add new records to the Registry. Each record has the following information (ID number, Full Name, Date of birth, Place of birth).
    - A user can search for someone using his ID number or name.<br><br>
Your program should handle all the exceptions that may occur like:
    - Searching for an ID number that doesn't exist
    - Wrong format for the date of birth

```python
import csv
import datetime

CR = None


def Operations():
    choice = 0
    global CR
    
    while (choice > 4 or choice < 1):
        
        choice = int(input("Enter number of operartion:\n 1-Initialize\n 2-Add new citizen\n 3-Search for citizen\n 4-Close file: "))
    
    print("")
    
    if choice == 1:
        Initial()
    if choice == 2:
        Resume()
        Append()
    if choice == 3:
        Open()
        Search()
    if choice == 4:
        CR.close()
        
def Initial():
    global CR
    try:
        CR = open(".\civilReg.csv", "w", newline='')
    except:
        print("Could not find or open the file")

    writer = csv.writer(CR)
    headings = ["ID Number", "Full Name", "Date of birth", "Place of birth"]
    writer.writerow(headings)
    
    Operations()


def Resume():
    global CR
    try:
        CR = open(".\civilReg.csv", "a", newline='')
    except:
        print("Could not find or open the file")
    

def Open():
    global CR
    try:
        CR = open(".\civilReg.csv", "r", newline='')
    except:
        print("Could not find or open the file")

        
def Append():
    global CR
    add = "Y"
    writer = csv.writer(CR)
        
    while(add == "Y" or add == "y"):
        name = input("Full name: ")
        ID = input("ID number: ")
        
        try:
            print("Date of Birth: ")
            date1 = input(datetime.date)
        except:
            print("Wrong date format")
        
        place = input("Place of birth: ")

        citizen  = [ID, name, date1, place]
        writer.writerow(citizen)
        add= input("Do you want to add another ? Y/N ")
    
    print("")
    Operations()

    
    
def Search():
    global CR
    reader = csv.reader(CR, delimiter=",")
    try:
        key = input("Enter person ID: ")
        for row in reader:
            
            if key == row[0]:
                 print(row )

    except:
        print("{} Couldn't be found", format(key))
    
    Operations()
    
## Start of Excution ##
Operations()




CR.close()
```
